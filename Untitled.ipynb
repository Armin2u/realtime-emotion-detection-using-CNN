# read the images and turn them into array

```python

training_Data = []

def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img))  # Load image
                if img_array is None:  # If image can't be read, skip
                    continue
                new_array = cv2.resize(img_array, (img_size, img_size))  # Resize image
                training_Data.append([new_array, class_num])  # Add to list
            except Exception as e:
                print(f"Error loading image {img}: {e}")
                continue  # Skip this image and continue with the next one
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
import random

random.shuffle(training_Data) #shouldnt learn the sequence
```

```python
x = [] #data
y = [] #label

for features,label in training_Data:
    x.append(features)
    y.append(label)

X = np.array(x).reshape(-1, img_size, img_size, 3) #4 dimentions
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((x, y))

# Shuffle, batch, and prefetch data for performance
train_dataset = train_dataset.shuffle(buffer_size=10000).batch(32).prefetch(tf.data.experimental.AUTOTUNE)

# Normalize images in the dataset
def normalize(image, label):
    image = tf.cast(image, tf.float32) / 255.0  # normalize
    return image, label

train_dataset = train_dataset.map(normalize)
```

# deep learning model to train. Transfer learning

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
model = Sequential()
```

```python
model.summary()
```

```python
# First layer: explicitly define input shape using Input() for the first layer
model.add(Input(shape=(64, 64, 3)))

# First convolutional layer with 32 filters and a 3x3 kernel, followed by max-pooling
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Second convolutional layer
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Third convolutional layer
model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the layers before passing to the fully connected layer
model.add(Flatten())

# Fully connected layer with 256 units
model.add(Dense(256, activation='relu'))

# Dropout to prevent overfitting
model.add(Dropout(0.5))

# Output layer (assuming you have 7 emotion categories, adjust if necessary)
model.add(Dense(7, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X, Y, epochs=10, batch_size=32, validation_split=0.2)
```

# for realtime detection

```python
import cv2
import numpy as np
import tensorflow as tf
import time
import sys

# Load the model
new_model = tf.keras.models.load_model("final_model_95p07.h5")
new_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Load face cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize webcam
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

last_status = None
last_update_time = time.time()
delay = 1  # Set delay time in seconds (adjust as needed)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    # Loop over the faces detected
    for (x, y, w, h) in faces:
        roi_color = frame[y:y+h, x:x+w]

        # Resize face region for model input to 64x64
        final_image = cv2.resize(roi_color, (64, 64))
        final_image = np.array(final_image)  # Ensure it's a NumPy array
        final_image = np.expand_dims(final_image, axis=0)  # Add batch dimension
        final_image = final_image / 255.0  # Normalize the image

        # Check the shape of final_image
        print("Input shape to the model:", final_image.shape)

        # Make predictions with the model
        Predictions = new_model.predict(final_image)

        # Determine the status (emotion)
        current_status = ""
        if np.argmax(Predictions) == 0:
            current_status = "Angry"
        elif np.argmax(Predictions) == 1:
            current_status = "Disgust"
        elif np.argmax(Predictions) == 2:
            current_status = "Fear"
        elif np.argmax(Predictions) == 3:
            current_status = "Happy"
        elif np.argmax(Predictions) == 4:
            current_status = "Sad"
        elif np.argmax(Predictions) == 5:
            current_status = "Surprised"
        else:
            current_status = "Neutral"

        # Check if the current status is different from the last status and if enough time has passed
        current_time = time.time()
        if current_status != last_status and (current_time - last_update_time) >= delay:
            last_status = current_status
            last_update_time = current_time  # Update the last update time

        # Display the last detected emotion on the frame
        if last_status is not None:  # Only display if there's a valid last status
            cv2.putText(frame, last_status, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Show the frame
    cv2.imshow('Face Emotion Recognition', frame)

    # Check if the window is closed or 'q' is pressed to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and destroy all windows
cap.release()
cv2.destroyAllWindows()
sys.exit()  # Exit the program
```
